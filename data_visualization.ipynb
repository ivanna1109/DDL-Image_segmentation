```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_from_pickle(filename):
    with open(filename, 'rb') as f:
        data = pickle.load(f)
    return data
```

```python
data = load_from_pickle('data.pkl')
if 'images' in data:
    print(f"Number of images: {len(data['images'])}")
if 'masks' in data:
    print(f"Number of masks: {len(data['masks'])}")
```

```python
df = pd.DataFrame.from_dict(data)
#print(df.head(5))
```

```python
def visualize_image_and_mask(image, mask, title=''):
    fig, axs = plt.subplots(1, 2, figsize=(8, 5))
    axs[0].imshow(image, cmap='gray')
    axs[0].set_title('Image')
    axs[0].axis('off')
    
    axs[1].imshow(mask, cmap='gray')
    axs[1].set_title('Mask')
    axs[1].axis('off')
    
    fig.suptitle(title)
    plt.show()
```

```python
i = 0
for idx, row in df.iterrows():
    if ( i < 5):
        image = row['images']
        mask = row['masks']
        visualize_image_and_mask(image, mask, title=idx)
        i+=1
```
